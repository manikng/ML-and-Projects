```python
import matplotlib.pyplot as plt

# Extract the year from the 'Subscription Date' column
df['year'] = pd.to_datetime(df['Subscription Date']).dt.year

# Group the data by year and country, and count the number of customers
customer_count = df.groupby(['year', 'Country']).size().reset_index(name='Count')

# Find the country with the highest number of customers
max_customers_country = customer_count.loc[customer_count['Count'].idxmax(), 'Country']

# Plotting the data
plt.figure(figsize=(12, 6))
plt.bar(customer_count['year'], customer_count['Count'], color='blue')
plt.xlabel('Year')
plt.ylabel('Number of Customers')
plt.title('Number of Customers Joined in Each Year')
plt.xticks(customer_count['year'])
plt.show()

print(f"The country with the highest number of customers is {max_customers_country}.")
```

```python
# Group the data by year and count the number of customers
customers_by_year = df.groupby('year').size().reset_index(name='Customers')

# Group the data by country and count the number of customers
customers_by_country = df.groupby('Country').size().reset_index(name='Customers')

# Find the country with the maximum number of customers
max_customers_country = customers_by_country.loc[customers_by_country['Customers'].idxmax(), 'Country']

# Print the number of customers in each year
print("Number of customers in each year:")
print(customers_by_year)

# Print the number of customers in each country
print("\nNumber of customers in each country:")
print(customers_by_country)

# Print the country with the maximum number of customers
print(f"\nThe country with the maximum number of customers is {max_customers_country}.")
```

```python
from sklearn.manifold import TSNE
model = TSNE(n_components=2, random_state=0)
tsne_data = model.fit_transform(df)
plt.figure(figsize=(7, 7))
plt.scatter(tsne_data[:, 0], tsne_data[:, 1])
plt.show()
```

```python
error = []
for n_clusters in range(1, 21):
	model = KMeans(init='k-means++',
				n_clusters=n_clusters,
				max_iter=500,
				random_state=22)
	model.fit(df)
	error.append(model.inertia_)
```

```python
plt.figure(figsize=(10, 5))
sb.lineplot(x=range(1, 21), y=error)
sb.scatterplot(x=range(1, 21), y=error)
plt.show()
```
